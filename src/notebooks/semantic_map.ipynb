```python
import json
import wizmap
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from config import CACHE_DIR
from utils import save_fig
import plotly.express as px
```

```python
# chunk-level
all_embeddings = []
all_text_chunks = []
chunk_category_labels = []
chunk_paper_labels = []
chunk_section_labels = []

# abstract-level
abstracts = []
abstract_embeddings = []

for paper in CACHE_DIR.iterdir():
    with open(paper, "r") as f:
        data = json.load(f)

    for chunk in data['body']:
        all_text_chunks.append(chunk['text'])
        all_embeddings.append(chunk['embedding'])
        chunk_category_labels.append(data['categories'][2])
        chunk_paper_labels.append(data['title'])
        chunk_section_labels.append(chunk['section'])
    
    for chunk in data['abstract']:
        abstracts.append(chunk['text'])
        abstract_embeddings.append(chunk['embedding'])

reducer = PCA(n_components=2)
embeddings_2d = reducer.fit_transform(all_embeddings)
abstract_embeddings_2d = reducer.fit_transform(abstract_embeddings)
```

**PCA Projection**

```python
hover_text = [f"<b>{title}</b><br>{section}<br>{text[:50]}..."
              for title, section, text in zip(chunk_paper_labels, chunk_section_labels, all_text_chunks)]

fig = px.scatter(
    x=embeddings_2d[:, 0],
    y=embeddings_2d[:, 1],
    hover_name=hover_text,
    color=chunk_category_labels,
    title="Chunk Embeddings",
    height=800
)
fig.show()
```

**WizMap**

```python
xs = embeddings_2d[:, 0].astype(float).tolist()
ys = embeddings_2d[:, 1].astype(float).tolist()

data_list = wizmap.generate_data_list(xs, ys, all_text_chunks)
grid_dict = wizmap.generate_grid_dict(xs, ys, all_text_chunks, "Paper Chunks")

wizmap.save_json_files(data_list, grid_dict, output_dir="./wizmap_json")

data_url = "http://localhost:8000/wizmap_json/data.ndjson"
grid_url = "http://localhost:8000/wizmap_json/grid.json"
```

```python
import http.server
import socketserver

PORT = 8000
DIRECTORY = "wizmap_json"

class CORSHandler(http.server.SimpleHTTPRequestHandler):
    def __init__(self, *args, **kwargs):
        super().__init__(*args, directory=DIRECTORY, **kwargs)

    def end_headers(self):
        self.send_header("Access-Control-Allow-Origin", "*")
        super().end_headers()

with socketserver.TCPServer(("", PORT), CORSHandler) as httpd:
    print(f"Serving at http://localhost:{PORT}")
    httpd.serve_forever()
```
